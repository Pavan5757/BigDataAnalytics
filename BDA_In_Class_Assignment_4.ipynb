```python
import pandas as pd

pd.set_option('display.max_rows', None)

student_data = pd.DataFrame({
    'school_code': ['s001', 's002', 's003', 's001', 's002', 's004'],
    'class': ['V', 'V', 'VI', 'VI', 'V', 'VI'],
    'name': ['Alberto Franco', 'Gino Mcneill', 'Ryan Parkes', 'Eesha Hinton', 'Gino Mcneill', 'David Parkes'],
    'date_of_birth': ['15/05/2002', '17/05/2002', '16/02/1999', '25/09/1998', '11/05/2002', '15/09/1997'],
    'age': [12, 13, 13, 14, 12, 11],
    'height': [173, 192, 186, 167, 151, 159],
    'weight': [35, 32, 33, 30, 31, 32],
    'address': ['street1', 'street2', 'street3', 'street1', 'street2', 'street4']
}, index=['S1', 'S2', 'S3', 'S4', 'S5', 'S6'])

print("Original DataFrame:")
print(student_data)
print("\nSplit the said data on school_code, class wise:")
result = student_data.groupby(['school_code', 'class'])
for name, group in result:
    print("\nGroup:")
    print(name)
    print(group)
```

```python
path = "/content/data.csv"
df = pd.read_csv(path)
```

```python
print("\nStatistical Description:")
print(df.describe())
```

```python
print("\nNull values:")
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
print("\nAggregated statistics (Duration and Calories):")
print(df[['Duration', 'Calories']].agg(['min', 'max', 'count', 'mean']))
```

```python
filtered_df1 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
print("\nCalories between 500 and 1000:")
print(filtered_df1)
```

```python
filtered_df2 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("\nCalories > 500 and Pulse < 100:")
print(filtered_df2)
```

```python
df_modified = df.drop(columns=['Maxpulse'])
```

```python
df.drop(columns=['Maxpulse'], inplace=True)
```

```python
df['Calories'] = df['Calories'].astype(int)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Duration'], df['Calories'], color='blue')
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title("Duration vs Calories")
plt.grid(True)
plt.show()
```
